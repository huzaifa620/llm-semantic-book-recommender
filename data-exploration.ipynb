```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv('books.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(df.isna().transpose(), cbar=False, ax=ax)

plt.xlabel('Columns')
plt.ylabel('Missing values')

plt.show()
```

```python
import numpy as np

df['missing_description'] = np.where(df['description'].isna(), 1, 0)
df['age_of_book'] = 2025 - df['published_year']

```

```python
column_of_interest = ['num_pages', 'age_of_book', 'missing_description', 'average_rating']

correlation_matrix = df[column_of_interest].corr(method='spearman')

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={"label": "Spearman Correlation"})
plt.show()
```

```python
book_missing = df[~(df['description'].isna()) & ~(df['num_pages'].isna()) & ~(df['average_rating'].isna()) & ~(df['published_year'].isna())]
```

```python
book_missing
```

```python
book_missing['categories'].value_counts().reset_index().sort_values('count', ascending=False)
```

```python
book_missing['words_in_description'] = book_missing['description'].apply(lambda x: len(str(x).split()))
```

```python
book_missing.loc[book_missing['words_in_description'].between(1, 5), 'description']
```

```python
book_missing_25_words = book_missing[book_missing['words_in_description'] >= 25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words['title_and_subtitle'] = (
    np.where(
        book_missing_25_words['subtitle'].isna(),
        book_missing_25_words['title'],
        book_missing_25_words['title'] + ": " + book_missing_25_words['subtitle']
    )
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words['tagged_description'] = book_missing_25_words[['isbn13', 'description']].astype(str).agg(' '.join, axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words.drop(['subtitle', 'missing_description', 'age_of_book', 'words_in_description'], axis=1).to_csv('books_cleaned.csv', index=False)
)
```
